```python
import numpy as np
```

```python
class Neuron1:
    def __init__(self):
        self.lr = 0.001
        self.w = 0.4
        self.bias = 0.2
        
    def train(self, input_x, target_y):
        length = len(input_x)
        for i in range(length):
            output_y = self.bias + self.w * input_x[i]
            error = target_y[i] - output_y
            self.bias += self.lr * error
            self.w += self.lr * error * input_x[i]

    
    def ask(self, input_x):
        output_y = self.bias*212 + self.w*input_x
        print(output_y)
```

```python
X = np.array([0, 100, 40, 20], dtype=float)   # Celsius
Y = np.array([32, 212, 104, 68], dtype=float) # Fahrenheit

X_norm = X / 212.0  # scale to ~0–1
Y_norm = Y / 212.0  # scale to ~0–1

nn = Neuron1()
```

```python
for epoch in range(10000):
    nn.train(X_norm, Y_norm)

print("Final bias (should be ~32):", nn.bias*212)
print("Final weight (should be ~1.8):", nn.w)

nn.ask(0)    # ~32
nn.ask(100)  # ~212
```
