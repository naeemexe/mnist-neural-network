```python
class Neuron1:
    def __init__(self):
        self.lr = 0.1
        self.m = 0.4
        
    def train(self, input_x, target_y):
        output_y = self.m*input_x
        error = target_y-output_y
        self.m += self.lr*error*input_x   #let Δm proportional to d(error)^2/dm
        print(self.m)                     #which equals Δw proportional to error*d(error)/dm
                                          #both represent the same thing
    def ask(self, input_x):
        output_y = self.m * input_x
        print(output_y)
```

```python
class Neuron2:
    def __init__(self):
        self.lr = 0.1
        self.m = 0.4
        
    def train(self, input_x, target_y):
        output_y = self.m*input_x
        error = target_y-output_y
        self.m += self.lr*error         #let Δw proportional to error only
        print(self.m)
        
    def ask(self, input_x):
        output_y = self.m*input_x
        print(output_y)
```

```python
m1 = Neuron1()
m2 = Neuron2()
```

```python
m1.train(1,0.62)
m1.train(2,1.24)
m1.train(1.2, 0.74)
m1.train(3, 1.86)
m1.train(2.3,1.42)
m1.train(4, 2.48) #overshoot
m1.train(0.4, 0.25) #more overshoot because of imperfect data
m1.train(2.6, 1.61)
m1.train(1.3, 0.81)
m1.train(3.7,2.29)
m1.train(4.8, 2.98)
```

```python
m2.train(1,0.62)
m2.train(2,1.24)
m2.train(0.5, 0.31)
m2.train(3, 1.86)
m2.train(2.1,1.30)
m2.train(1, 0.62)
m2.train(4, 2.48)
m2.train(0.4, 0.25)
m2.train(4.2, 2.60)
m2.train(2.9, 1.8)
m2.train(3.1, 1.92)
m2.train(2.5, 1.55)
m2.train(1.1, 0.68)
m2.train(3.6,2.23)
m2.train(4.6, 2.85)
m2.train(0.2, 0.12)
```
